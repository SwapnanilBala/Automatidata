### Boxplots

**Let us perform a check for outliers on relevant columns such as trip distance and trip duration. We should remember, some of the best ways to identify the presence of outliers in data are `box plots` and `histograms`.** 

**Note:** Remember to convert your date columns to datetime in order to derive total trip duration.  


### ðŸ§© Let's start with converting datatypes of targeted columns to useful ones and then lets plot something


```python
# Converting data columns to datetime
### OUR CODE ###
df['tpep_pickup_datetime'] = pd.to_datetime(df['tpep_pickup_datetime']) 
# you could learn more about the `to_datetime` from my pandas practice library
df['tpep_dropoff_datetime'] = pd.to_datetime(df['tpep_dropoff_datetime'])
```

```python
# Creating box plot of trip_distance
### OUR CODE ###
plt.figure(figsize=(8,3)) # Adjusts the size of the figure
plt.title('Trip_distance') # This is how we choose the name of the plot
sns.boxplot(data=None, x=df['trip_distance'], fliersize = 1); # make sure not to confuse the names of the target column
```

#### *Did you see that there were so many outliersðŸŽ² the right whisker ended at around 5 but man the outliers extended to almost 35ðŸ˜…*

```python
# Creating histogram of trip_distance
### OUR CODE ###
plt.figure(figsize=(10,5))
sns.histplot(df['trip_distance'], bins = range(0,26,1))
plt.title('Trip Distance Histogram');
```

### ðŸ•¶ï¸ So Here, the counts stands for how many trips have been made with regards to the trip distance

```python
# Creating a  box plot of total_amount
### OUR CODE ###
plt.figure(figsize=(7,2))
sns.boxplot(data= None, x = df['total_amount'], fliersize = 1)
plt.title('total amount')
```

**Mean tips by passenger count**

Let us examine the unique values in the `passenger_count` column.

```python
### OUR CODE ###
df['passenger_count'].value_counts()
```

```python
# Calculating mean tips by passenger_count

### OUR CODE ###
df[['passenger_count','tip_amount']].groupby('passenger_count').mean()[['tip_amount']]

```

```python
# Creating a bar plot for mean tips by passenger count, its a complicated one, let us take some time to understand it fully

### OUR CODE ###
mean_tips_by_passenger_count = (
    df.groupby('passenger_count', as_index=False)['tip_amount']
      .mean()
      .dropna()
)

plt.figure(figsize=(11,4))
ax = sns.barplot(
    x='passenger_count',
    y='tip_amount',
    data=mean_tips_by_passenger_count,
    palette='Greens_d',
    
)
ax.axhline(df['tip_amount'].mean(), ls='--', color='red', label='Global Mean')
ax.legend()
plt.title('Mean Tip Amount by Passenger Count', fontsize=28)
plt.xlabel('Passenger Count')
plt.ylabel('Mean Tip Amount')
plt.show()
```

**Creating month and day columns**

```python
# Create a month column

### OUR CODE ###
df['month'] = df['tpep_dropoff_datetime'].dt.month_name()

# You may find pandas tutorial in my other repository

# Create a day column

### OUR CODE ###
df['day'] = df['tpep_dropoff_datetime'].dt.day_name()
```

**Plot total ride count by month**

Let us begin by calculating total ride count for every month.

```python
# Get total number of rides for each month
### OUR CODE ###
number_of_rides_per_month = df['month'].value_counts()
number_of_rides_per_month
```

Reorder the results to put the months in calendar order.

```python
# Reorder the monthly ride list so months go in order
### OUR CODE ###
month_order = ['January', 'February', 'March', 'April', 'May', 'June', 'July',
         'August', 'September', 'October', 'November', 'December']

monthly_ride = number_of_rides_per_month.reindex(index = month_order)
monthly_ride
```

```python
# Showing the index

### OUR CODE ###
monthly_ride.index
```

```python
# Creating a bar plot of total rides per month
### OUR CODE ###
plt.figure(figsize=(13,6))
ax = sns.barplot(x = monthly_ride.index, y = monthly_ride)
ax.set_xticklabels(month_order)
plt.ylabel('Rides')
plt.title('Total Rides Per Month', fontsize = 24);
```

**Plot total ride count by day**

Repeat the above process, but now calculate the total rides by day of the week.

```python
# Repeating the above process, this time for rides by day
### OUR CODE ###
number_of_rides_per_day = df['day'].value_counts()
day_order = ['Sunday', 'Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday']
day_ordered = number_of_rides_per_day.reindex(index = day_order)

plt.figure(figsize=(13,7))
ax = sns.barplot(x = day_ordered.index, y = day_ordered )
ax.set_xticklabels(day_order)
ax.set_ylabel('Number of Rides')
plt.title('Rides by Day', fontsize = 20);
```

**Plotted the total revenue by day of the week**

Repeat the above process, but now calculate the total revenue by day of the week.

```python
# Repeating the process, this time for total revenue by day
### OUR CODE ###

total_amount_day = df[['day','total_amount']].groupby('day').sum()[['total_amount']]
total_amount_day = total_amount_day.reindex(index=day_order)
total_amount_day
```

```python
# Create bar plot of total revenue by day
### OUR CODE ###
plt.figure(figsize=(12,7))
ax = sns.barplot(x=total_amount_day.index, y=total_amount_day['total_amount'])
# as we used groupby function we needed to mention the specific column of the dataset
ax.set_xticklabels(day_order)
ax.set_ylabel('Revenue (USD)')
plt.title('Total revenue by day', fontsize=16);
```

```python
# Scatterplot isn't ideal for this case, still we will try out 

plt.figure(figsize=(13,7))
sns.scatterplot(
    x=day_ordered.index,
    y=day_ordered.values,
    s=100,                   # dot size
    color='teal',
    edgecolor='black'
)
plt.ylabel('Number of Rides')
plt.title('Rides by Day (Scatter Representation)', fontsize=20)
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

**Plot total revenue by month**

```python
# Repeating the process, this time for total revenue by month
### OUR CODE ###
revenue_by_month = df[['month','total_amount']].groupby('month').sum()[['total_amount']]
revenue_by_month = revenue_by_month.reindex(index = month_order)

plt.figure(figsize=(13,8))
ax = sns.barplot(x = revenue_by_month.index, y = revenue_by_month['total_amount'])
ax.set_ylabel('Revenue (USD)')
plt.title('Total Revenue By Month', fontsize = 24);
```

## BONUS CONTENT

To confirm your conclusion, consider the following experiment:
1. Create a sample of coordinates from a normal distribution&mdash;in this case 1,500 pairs of points from a normal distribution with a mean of 10 and a standard deviation of 5
2. Calculate the distance between each pair of coordinates 
3. Group the coordinates by endpoint and calculate the mean distance between that endpoint and all other points it was paired with
4. Plot the mean distance for each unique endpoint

```python
#BONUS CONTENT

#1. Generating random points on a 2D plane from a normal distribution
### OUR CODE ###

x = np.random.normal(loc = 0, scale = 1, size = 500)
y = np.random.normal(loc = 0, scale = 1, size = 500)

random_points = np.column_stack((x,y))
print(random_points.shape)

# 2. Calculating Euclidean distances between points in first half and second half of array
### OUR CODE ###

half = len(random_points) // 2
first_half = random_points[:half]
last_half = random_points[half:]
print(f'Here is the size of the first half: {first_half.shape}, and the other half: {last_half.shape}')

Eucli_Dis = np.sqrt(((first_half[:, np.newaxis, :] - last_half[np.newaxis, :, :])**2).sum(axis=2))
print(Eucli_Dis.shape)
# 3. Grouping the coordinates by "drop-off location", computing the mean distance
### OUR CODE ###

df[['DOLocationID', 'trip_distance']].groupby('DOLocationID').mean()['trip_distance']

# 4. Plot the mean distance between each endpoint ("drop-off location") and all points it connected to
### OUR CODE ###
mean_distance_per_point = Eucli_Dis.mean(axis=1)  # mean distance of each first-half point

plt.figure(figsize=(10,6))
plt.plot(mean_distance_per_point, color='teal')
plt.title('Mean Euclidean Distance for Points in First Half')
plt.xlabel('Point Index (First Half)')
plt.ylabel('Mean Distance to Second Half')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
